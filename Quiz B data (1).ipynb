```python
#Normalizing the dataset to a common scale 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
normalised_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
features_df = normalised_df.drop(columns=['Appliances'])
Appliances = normalised_df['Appliances'] 
normalised_df.head()


```

```python
# split the dataset into training and testing dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df,Appliances, test_size=0.3, random_state=1)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train,y_train) # fit the model to the training dataset
linear_predict = linear_model.predict(x_test) # obtains predictions
linear_model.coef_
#sns.regplot(x=y_test, y=predicted_values)
```

```python
#Ridge regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
Ridge_predict = ridge_reg.predict(x_test)
ridge_reg.coef_
```

```python
#lasso regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train,y_train) 
lasso_predict=lasso_reg.predict(x_test)
lasso_reg.coef_
```

```python
# Comparing the effects of Regularization
def get_weights_df(model,feat,col_name):
    weights = pd.Series(model.coef_,feat.columns).sort_values()
    weights_df=pd.DataFrame(weights).reset_index()
    weights_df.columns=['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights =pd.merge(final_weights, lasso_weights_df, on='Features')


final_weights
```

```python
# Absolute Error
from sklearn.metrics import mean_absolute_error
mae_linear = mean_absolute_error(y_test, linear_predict)
mae_lasso= mean_absolute_error(y_test, lasso_predict)
mae_Ridge = mean_absolute_error(y_test, Ridge_predict)
mae_linear =round(mae_linear,3)
mae_lasso =round(mae_lasso,3)
mae_Ridge =round(mae_Ridge,3)
print("Absolute error for linear reg=", mae_linear)
print("Absolute error for lasso reg=", mae_lasso)
print("Absolute error for Ridge=", mae_Ridge)
```

```python
#R squared value
from sklearn.metrics import r2_score
r2_linear =r2_score(y_test,linear_predict)
r2_lasso =r2_score(y_test,lasso_predict)
r2_ridge =r2_score(y_test,Ridge_predict)
r2_linear=round(r2_linear,3)
r2_lasso=round(r2_lasso,3)
r2_ridge=round(r2_ridge,3)
print("R square value for linear reg=", r2_linear)
print("R square value for lasso reg=", r2_lasso)
print("R square value for Ridge reg=", r2_ridge)
```

```python
#residual sum of squares for linear regression
rss = np.sum(np.square(y_test - linear_predict))
round(rss,2)
```

```python
#root mean squared error for linear regression
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, linear_predict))
round(rmse,3)
```

```python
#root mean squared error for ridge regression
rmse1 = np.sqrt(mean_squared_error(y_test, Ridge_predict))
round(rmse1,3)
```

```python
#root mean squared error for lasso regression
rmse1 = np.sqrt(mean_squared_error(y_test, lasso_predict))
round(rmse1,3)
```
